```
import numpy as np
from sklearn.model_selection import learning_curve
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Generate simulated regression data
np.random.seed(42)
X_regression = np.random.rand(100, 1)  # Input feature
y_regression = 2 * X_regression.squeeze() + np.random.randn(100)  # Target variable

def plot_learning_curves(X_regression, y_regression, regression_estimator, title):
    """
    Plot the learning curves for a regression task.

    Parameters:
    - X_regression (numpy.ndarray): Input features for regression.
    - y_regression (numpy.ndarray): Target variable for regression.
    - regression_estimator: Regression estimator object.
    - title (str): Title for the plot.

    Returns:
    - None
    """

    train_sizes, train_scores, test_scores = learning_curve(regression_estimator, X_regression, y_regression, cv=5)
    
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)
    
    plt.figure()
    plt.title(title)
    plt.xlabel("Training Examples")
    plt.ylabel("Score")
    plt.grid()
    
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_mean, 'o-', color="r", label="Training Score")
    plt.plot(train_sizes, test_mean, 'o-', color="g", label="Cross-validation Score")
    plt.legend(loc="best")
    plt.show()

# Regression Learning Curve
regression_estimator = LinearRegression()
plot_learning_curves(X_regression, y_regression, regression_estimator, "Regression Learning Curve")
```
